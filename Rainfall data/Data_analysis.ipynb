# **Data: Sub Divisional Monthly Rainfall from 1901 to 2017**

# Import dataset and print the statistical info

```python
import pandas as pd

df = pd.read_csv("Sub_Division_IMD_2017.csv")
df.info()
df.describe()
```

```python
df.head()
```

# Data Cleaning

```python
df = df.replace("NA", pd.NA)
df.iloc[:,2:] = df.iloc[:,2:].apply(pd.to_numeric, errors='coerce')
```

```python
df.isna().sum()
```

```python
# Here we replace each NaN value with mean of the corresponding column
df.fillna(df.select_dtypes(include='number').mean(), inplace=True)
```

# Basic Data Statistics

```python
df['ANNUAL'].describe()
```

```python
# We'll print the unique entries of 'SUBDIVISION'

unique_subdivisions = df['SUBDIVISION'].unique()
print(unique_subdivisions)
```

```python
# Now we'll divide the given regions into 8 categories: N, E, W, S, NE, SE, SW, C

import matplotlib.pyplot as plt
region_map = {
    # North
    'Jammu & Kashmir': 'N', 'Himachal Pradesh': 'N',
    'Punjab': 'N', 'Haryana Delhi & Chandigarh': 'N', 'Uttarakhand': 'N',

    # North-East
    'Arunachal Pradesh': 'NE', 'Assam & Meghalaya': 'NE',
    'Naga Mani Mizo Tripura': 'NE', 'Sub Himalayan West Bengal & Sikkim': 'NE',

    # East
    'Gangetic West Bengal': 'E', 'Orissa': 'E',
    'Jharkhand': 'E', 'Bihar': 'E',

    # Central
    'East Uttar Pradesh': 'C', 'West Uttar Pradesh': 'C',
    'East Madhya Pradesh': 'C', 'West Madhya Pradesh': 'C',
    'Chhattisgarh': 'C',

    # West
    'West Rajasthan': 'W', 'East Rajasthan': 'W',
    'Gujarat Region': 'W', 'Saurashtra & Kutch': 'W',

    # South
    'Tamil Nadu': 'S', 'Rayalseema': 'S', 'South Interior Karnataka': 'S',
    'Kerala': 'S', 'Lakshadweep': 'S',

    # South-West
    'Konkan & Goa': 'SW', 'Madhya Maharashtra': 'SW',
    'Matathwada': 'SW', 'Vidarbha': 'SW',
    'Coastal Karnataka': 'SW', 'North Interior Karnataka': 'SW',

    # South-East
    'Coastal Andhra Pradesh': 'SE', 'Telangana': 'SE',
    'Andaman & Nicobar Islands': 'SE',
}

# Add REGION column
df["REGION"] = df["SUBDIVISION"].map(region_map)

# Next we'll plot region-wise annual rainfall trend in 8 sub-plots
regions = ["N","NE","E","C","W","S","SE","SW"]
fig, axes = plt.subplots(4, 2, figsize=(18, 12), sharex=True, sharey=False)
axes = axes.flatten()

for i, region in enumerate(regions):
    ax = axes[i]
    sub_df = df[df["REGION"] == region]
    if sub_df.empty:
        ax.set_title(region + " (No Data)")
        continue

    for subdivision, group in sub_df.groupby("SUBDIVISION"):
        ax.plot(group["YEAR"], group["ANNUAL"], label=subdivision)

    ax.set_title(f"{region} Region")
    ax.set_xlabel("Year")
    ax.set_ylabel("Annual Rainfall (mm)")
    ax.legend(fontsize=7, loc="best")

plt.tight_layout()
plt.show()
```

# Analysis using plots

**Scatter Plot**

```python
# Region-wise scatter plot of annual rainfall VS JJAS rainfall to see contribution of monsoon in annual rainfall

fig, axes = plt.subplots(2, 4, figsize=(18, 9), sharex=True, sharey=True)
axes = axes.flatten()
for i, region in enumerate(regions):
    ax = axes[i]
    sub_df = df[df["REGION"] == region]
    if sub_df.empty:
        ax.set_title(region + " (No Data)")
        continue
    for subdivision, group in sub_df.groupby("SUBDIVISION"):
        ax.scatter(group["JJAS"], group["ANNUAL"], label=subdivision, alpha=0.6)
    ax.set_title(f"{region} Region")
    ax.set_xlabel("JJAS Rainfall (mm)")
    ax.set_ylabel("Annual Rainfall (mm)")
    ax.legend(fontsize=6, loc="best")

plt.suptitle("Annual vs JJAS Rainfall by Region", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

**Observations**:
1. Strong positive linear relationship indicates that JJAS (monsoon months) contribute the major portion of the annual rainfall across all regions.
2. In North-eastern regions, more spread/scatter is observed which indicates that rainfall outside JJAS months also plays a vital role.
3. Some South-East and South regions (like Andaman & Nicobar, Tamil Nadu, Rayalaseema) show deviations which is explained by their dependency on Northeast Monsoon (OctDec).


**Box Plot**

```python
# Here, we consider 1955 as the beginning of computer-era in India and compare season wise rainfall for pre- VS post- computer era

import seaborn as sns

df["Era"] = df["YEAR"].apply(lambda x: "Pre-Computer (<=1955)" if x <= 1955 else "Post-Computer (>1955)")
seasonal_cols = ["JF", "MAM", "JJAS", "OND"]
df_melt = df.melt(
    id_vars=["YEAR", "REGION", "Era"],
    value_vars=seasonal_cols,
    var_name="Season",
    value_name="Rainfall"
)

g = sns.catplot(
    data=df_melt, kind="box",
    x="Season", y="Rainfall",
    hue="Era", col="REGION",
    col_wrap=4, sharey=False,
    palette="Set2", height=4
)
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle("Seasonal Rainfall by Region: Pre- vs Post-Computer Era", fontsize=16)
plt.show()
```

**Observations:**
1. Most regions show considerable variability in rainfall across seasons with JJAS(monsoon months) taking up maximum rainfall value.
2. Post-Computer era shows slightly lower median rainfall in many regions but has more extreme high rainfall outliers in some regions (like NE, SW, S).
3. Variability is observed to change regionally. Some regions have narrower post-computer rainfall distribution (such as SE, N), while others (such as NE, SW, S) see more spread and extreme events.

**Bar Plot**

```python
# Here we compare the rainfall pattern of decade 1901-1910 and 2008-2017 for the unique regional entries of 'SUBDIVISION' column.

decade1 = (1901, 1910)
decade2 = (2008, 2017)

# Compute mean annual rainfall per SUBDIVISION for each decade
df_decade1 = df[(df['YEAR'] >= decade1[0]) & (df['YEAR'] <= decade1[1])]
df_decade2 = df[(df['YEAR'] >= decade2[0]) & (df['YEAR'] <= decade2[1])]

mean_decade1 = df_decade1.groupby('SUBDIVISION')['ANNUAL'].mean().reset_index()
mean_decade2 = df_decade2.groupby('SUBDIVISION')['ANNUAL'].mean().reset_index()
mean_decade1 = mean_decade1.sort_values('ANNUAL')
mean_decade2 = mean_decade2.sort_values('ANNUAL')

# Plot for first decade
plt.figure(figsize=(12,6))
sns.barplot(data=mean_decade1, x='SUBDIVISION', y='ANNUAL', palette="Blues_d")
plt.xticks(rotation=90)
plt.title('Mean Annual Rainfall by Subdivision (1901-1910)')
plt.ylabel('Mean Annual Rainfall (mm)')
plt.xlabel('Subdivision')
plt.tight_layout()
plt.show()

# Plot for second decade
plt.figure(figsize=(12,6))
sns.barplot(data=mean_decade2, x='SUBDIVISION', y='ANNUAL', palette="Greens_d")
plt.xticks(rotation=90)
plt.title('Mean Annual Rainfall by Subdivision (2008-2017)')
plt.ylabel('Mean Annual Rainfall (mm)')
plt.xlabel('Subdivision')
plt.tight_layout()
plt.show()
```

**Observations:**
1. Coastal Karnataka, Andaman & Nicobar Islands, Kerala, and Konkan & Goa consistently receive highest rainfall in both periods. Similarly, West Rajasthan and Saurashtra & Kutch remain the driest regions.
2. Some regions (e.g., Gangetic West Bengal, Jharkhand, Bihar) appear to have decreased in ranking in terms of rainfall. Chhattisgarh and Orissa also appear to have slightly lower values or rankings in the second period. This suggests possible drying trends in eastern-central India.
3. Western Ghats regions (like Konkan) have slightly increased rainfall in 2008-2017 decade.
4. The middle tier (e.g., Telangana, East Madhya Pradesh, Madhya Maharashtra) shows shuffling in ranks, indicating local changes in rainfall patterns

**Heatmap**

```python
# Next we plot the heatmap between seasonal periods: JF, MAM, JJAS, OND and Annual

seasonal_cols = ["JF", "MAM", "JJAS", "OND", "ANNUAL"]

plt.figure(figsize=(8,6))
sns.heatmap(df[seasonal_cols].corr(), annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Heatmap of Seasonal & Annual Rainfall (National Level)", fontsize=14)
plt.show()
```

```python
# Heatmap to show Regional Mean Seasonal Rainfall Correlation

regional_means = df.groupby(["YEAR", "REGION"])[seasonal_cols].mean().reset_index()
# Pivot to wide form: each region as column, values = seasonal rainfall
for season in seasonal_cols:
    regional_wide = regional_means.pivot(index="YEAR", columns="REGION", values=season)
    plt.figure(figsize=(8,6))
    sns.heatmap(regional_wide.corr(), annot=True, cmap="coolwarm", vmin=-1, vmax=1)
    plt.title(f"Regional Correlation of {season} Rainfall", fontsize=14)
    plt.show()
```

**Observations:**
1. SW-W, C-N and SW-S pairs show strong positive correlation, illustrating similar rainfall patterns.
2. In annual rainfall, NE-W pair has the most negative correlation portraying  vastly different rainfall behavior.
3. NE is an outlier with minimal similarity to others.
